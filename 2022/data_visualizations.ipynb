```python
import matplotlib.pyplot as plt
import numpy as np

import day17
import day18

USE_EXAMPLE_FILES = True
# USE_EXAMPLE_FILES = False

if USE_EXAMPLE_FILES:
    FILE17_NAME = "puzzle_inputs/day17_example.txt"
    FILE18_NAME = "puzzle_inputs/day18_example.txt"
else:
    FILE17_NAME = "puzzle_inputs/day17_example.txt"
    FILE18_NAME = "puzzle_inputs/day18_example.txt"
```

```python
TOP_DISPLAYED_HEIGH = 130
LOWER_DISPLAYED_HEIGH = 0

with open(FILE17_NAME) as file:
    jet_pattern = file.read().splitlines()[0]
blocked_positions = np.array(list(day17.solve_first_part(jet_pattern, print_result=False)))


rock_plot = np.zeros([day17.CAVE_WIDENESS, TOP_DISPLAYED_HEIGH
 - LOWER_DISPLAYED_HEIGH + 1], dtype= bool)

rocks_in_plot = blocked_positions[ (blocked_positions[:,1] < TOP_DISPLAYED_HEIGH
) & (blocked_positions[:,1] >= LOWER_DISPLAYED_HEIGH) ]
rocks_in_plot[:,1] -= LOWER_DISPLAYED_HEIGH

for rocks in rocks_in_plot:
    rock_plot[ tuple(rocks) ] = True

plt.figure(figsize=(1,7))
plt.imshow(rock_plot.T, origin='lower', cmap="winter", extent=[0,day17.CAVE_WIDENESS,LOWER_DISPLAYED_HEIGH,TOP_DISPLAYED_HEIGH
]);
# plt.savefig('day17_rock_plot.png', dpi=200)
```

```python
with open(FILE18_NAME) as file:
    lines18 = file.read().splitlines()
while(lines18[-1] == ""): # remove last empty lines, if any. They do not add information and can cause confusion
    lines18.pop()

boulder = day18.parse_puzzle_file(lines18)

%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

alpha = 1
cube_colors = np.ones([boulder.shape[0],boulder.shape[1],boulder.shape[2],4]) * [0.4,0.3,0.3,1] # red, green, blue, alpha (all normalized)
ax.voxels(boulder, facecolors=cube_colors);
```
